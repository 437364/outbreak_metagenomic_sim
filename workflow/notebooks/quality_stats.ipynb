```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
multiqc_data_path = "results/multiqc/multiqc_data/multiqc_general_stats.txt"
df=pandas.read_csv(multiqc_data_path, sep="\t")
# rename columns
# Sample  FastQC_mqc-generalstats-fastqc-percent_duplicates       FastQC_mqc-generalstats-fastqc-percent_gc       FastQC_mqc-generalstats-fastqc-avg_sequence_length      FastQC_mqc-generalstats-fastqc-median_sequence_length        FastQC_mqc-generalstats-fastqc-percent_fails    FastQC_mqc-generalstats-fastqc-total_sequences
df.columns = ["Sample", "Percent Duplicates", "Percent GC", "Avg Sequence Length", "Median Sequence Length", "Percent Fails", "Total Sequences"]
df["Total Sequences (millions)"] = df["Total Sequences"] / 1000000
```

```python
min_reads = np.min(df["Total Sequences"])
max_reads = np.max(df["Total Sequences"])
print(f"Min reads: {min_reads}")
print(f"Max reads: {max_reads}")
```

```python
# vertical boxplot of the number of reads per sample, each sample is one dot
plt.figure(figsize=(3, 5))
sns.boxplot(data=df, y="Total Sequences (millions)", showfliers=False, color="#3768c0")
sns.swarmplot(data=df, y="Total Sequences (millions)", color="black")
plt.ylabel("Total sequences (millions)")
plt.savefig("results/total_sequences_boxplot.svg")
```

```python
# boxplot of the percent of duplicates per sample
plt.figure(figsize=(3, 5))
sns.boxplot(data=df, y="Percent Duplicates", showfliers=False, color="#3768c0")
sns.swarmplot(data=df, y="Percent Duplicates", color="black")
plt.ylabel("Duplicates (%)")
plt.show()
```

```python
# boxplot of the read length per sample
plt.figure(figsize=(3, 5))
sns.boxplot(data=df, y="Avg Sequence Length", showfliers=False, color="#3768c0")
sns.swarmplot(data=df, y="Avg Sequence Length", color="black")
plt.ylabel("Avg Sequence Length")
plt.show()
```
